```python
# 1. 데이터 로드
path_Wilcoxon_DEG_Myeloid = "/nas/data/arpa_h/Myeloid_cell/Wilcoxon_DEG_Myeloid.csv"
df_wilcoxon = pd.read_csv(path_Wilcoxon_DEG_Myeloid, index_col=0)  # Unnamed: 0 → gene names
df_wilcoxon.reset_index(inplace=True)
df_wilcoxon.rename(columns={'index': 'name'}, inplace=True)
df_wilcoxon = df_wilcoxon[["names", "logfoldchanges", "pvals", "pvals_adj"]]
df_wilcoxon.rename(columns={"names": "gene","pvals": "pval","pvals_adj": "Wilcoxon_adj_pval"}, inplace=True)
df_wilcoxon
```
